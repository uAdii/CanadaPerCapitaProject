```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('PCI')
plt.scatter(x=df.year,y=df['per capita income (US$)'],color='blue',marker='+')
```

```python
new_df = df.drop('per capita income (US$)', axis='columns')
new_df
```

```python
PCI= df['per capita income (US$)']
PCI
```

```python
#CREATING LINEAR REGRESSION
reg = linear_model.LinearRegression()
reg.fit(new_df,PCI)
```

```python
#PREDICTING PER CAPITA INCOMES
reg.predict([[2020]])
```

```python
#slope of the line
reg.coef_
```

```python
#intercept of the graph
reg.intercept_
```

```python
reg.predict([[2021]])
```
